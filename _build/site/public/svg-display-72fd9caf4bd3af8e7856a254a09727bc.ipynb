# SVG display

De code hieronder kan gebruikt worden om SVG-code te tonen.
De invoer kan gegeven worden in de cell met `svgimage`.
Door de volgende cellen uit te voeren wordt de figuur verder opgebouwd en getoond.

````python
from IPython.display import display, HTML
display(HTML('<h1>Hello, world!</h1>'))
````

Een voorbeeld van een svg-figuur

````python
mysvg = '''
<svg width="300" height="300" xmlns="http://www.w3.org/2000/svg">
<g transform="translate(10.0 10.0)" stroke="black" fill="white"> 
<circle cx="50.0" cy="50.0" r="20.0"  /> 
<rect x="50.0" y="10.0" w="10.0" h="20.0"  />
</g>'''
````

````python
display(HTML(mysvg))
````

````python
svgimage = '''
<circle cx=50.0 cy=50.0 r=20.0 /> 
'''
````

````python
svgframe = '<svg width="400" height="200" xmlns="http://www.w3.org/2000/svg">' + \
           '<g fill="white" stroke="black" stroke-width="2">' + \
             svgimage + \
           '</g>' + \
           '</svg>'
````

````python
display(HTML(svgframe))
````
