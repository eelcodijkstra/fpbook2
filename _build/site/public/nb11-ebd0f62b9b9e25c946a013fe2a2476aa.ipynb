# Bomen

```haskell
import Data.Char
```

Op eenzelfde manier als we lijsten gedefinieerd hebben, kunnen we ook *bomen* definiëren. We beginnen met *binaire bomen*. Later breiden we dit uit tot n-aire bomen.

We definiëren een *generieke* binaire boom: de elementen in de boom moeten van eenzelfde type zijn, maar dat kan verder een willekeurig type zijn.

```haskell
data Tree a = Nil | Node a (Tree a) (Tree a) deriving (Show)
```

```haskell
tmap :: (a -> b) -> (Tree a) -> (Tree b)
tmap f Nil = Nil
tmap f (Node x lt rt) = Node (f x) (tmap f lt) (tmap f rt)
```

```haskell
tree1 = Node 1 (Node 2 Nil Nil) (Node 3 (Node 4 Nil Nil) Nil)
```

```haskell
tree2 = Node "aap" (Node "noot" (Node "mies" Nil Nil) Nil) (Node "teun" (Node "vuur" Nil Nil) Nil)
```

```haskell
tmap (\x -> x*x) tree1
```

```haskell
upper :: [Char] -> [Char]
upper [] = []
upper (x:xs) = (toUpper x) : (upper xs)
```

```haskell
upper "aap"
```

```haskell
tmap upper tree2
```

**Flatten** We kunnen een boom "platslaan": we plaatsen alle knoop-elementen dan in een lijst. 

We kunnen een boom op verschillende manieren platslaan: prefix, infix, of postfix. We gebruiken eerst de "prefix", waarbij de volgorde van de knopen gelijk blijft aan die in de definitie (die ook prefix is).

```haskell
flatten :: (Tree a) -> [a]
flatten Nil = []
flatten (Node x lt rt) = x : ((flatten lt) ++ (flatten rt))
```

```haskell
flatten tree1
```

```haskell
flatten tree2
```
