# SVG display

Dit is een Python-notebook, bedoeld om SVG-figuren weer te geven, met behulp van de HTML-display mogelijkheden.

````python
from IPython.display import display, HTML
````

Een voorbeeld van het tonen van HTML:

````python
display(HTML('<h3>Hello, world!</h3>'))
````

Een voorbeeld van een svg-figuur

````python
mysvg = '''
<svg width="300" height="200" xmlns="http://www.w3.org/2000/svg">
<g transform="translate(10.0 10.0)" stroke="black" fill="white"> 
<circle cx="50.0" cy="50.0" r="20.0"  /> 
<rect x="50.0" y="50.0" width="100.0" height="20.0" " />
</g>'''
````

````python
display(HTML(mysvg))
````

Een SVG-figuur heeft een context nodig, deze kun je eventueel aanpassen, bijvoorbeeld om een andere vulkleur te krijgen, of een andere lijn-kleur of -dikte.

````python
svgframe = '<svg width="400" height="200" xmlns="http://www.w3.org/2000/svg">' + \
           '<g fill="white" stroke="black" stroke-width="1">' + \
             svgimage + \
           '</g>' + \
           '</svg>'
````

````python
display(HTML(svgframe))
````
